```python
#Funcio per asignar una gen a cada joc de pokemon
def asignar_generacion(df):
    Gen = {
        'Red' : 'Gen1',
        'X' : 'Gen6',
        'Sword' : 'Gen8',
        'Omega Ruby' : 'Gen6',
        'Sun' : 'Gen7',
        'Let\'s Go Pikachu' : 'Gen7',
        'Legends Arceus' : 'Gen8',
        'Scarlet' : 'Gen9',
        'Gold' : 'Gen2',
        'Ruby' : 'Gen3',
        'Fire Red' : 'Gen3',
        'Leaf Green' : 'Gen3',
        'Emerald' : 'Gen3',
        'Diamond' : 'Gen4',
        'Platinum' : 'Gen4',
        'Black' :  'Gen5',
        'Black 2' : 'Gen5',
        'Ultra Sun' : 'Gen7',
        'Pokémon GO' : 'Gen7',
        'Violet' : 'Gen9',
    }
    df['Gen'] = df['Game(s) of Origin'].apply(lambda x: Gen.get(x.strip('"').strip(), 'Generación desconocida'))    
    return df
```

```python
#Executar la funcio
pokedb = asignar_generacion(pokedb)
```
